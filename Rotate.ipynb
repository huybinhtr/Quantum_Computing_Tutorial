## Operation of Roation operators R(x,y,z) on the ground state

```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
q = QuantumRegister(1)
c = ClassicalRegister(1)
qc = QuantumCircuit(q,c)
```

```python
qc.draw('mpl')
```

```python
from qiskit.quantum_info import Statevector
```

```python
psi1 = Statevector(qc)
psi1.draw('latex')
```

```python
from qiskit.visualization import plot_state_qsphere
```

```python
plot_state_qsphere(qc)
```

```python
from qiskit.quantum_info import DensityMatrix
```

```python
rho1 = DensityMatrix(qc)
rho1.draw('latex',prefix='\\rho_1 = ')
```

```python
from qiskit.visualization import plot_state_city
plot_state_city(qc)
```

```python
from qiskit.visualization import plot_state_hinton
plot_state_hinton(qc)
```

```python
from qiskit.visualization import plot_state_paulivec
plot_state_paulivec(qc,title='Initial Density Matrix as linear combination of Pauli gates')
```

```python
from qiskit.visualization import plot_bloch_multivector
plot_bloch_multivector(qc)
```

```python
purity = rho1.purity()
print(purity)
```

```python
from math import pi 
```

```python
qc.rx(pi/2,q[0]) # Rotation R(x,y,z) and has to define an angle pi/2 for x & z and pi/3 for y)
```

```python
qc.draw('mpl')
```

```python
psi2 = Statevector(qc)
psi2.draw('latex')
```

```python
plot_state_qsphere(qc) #state 0 & 1 has the same probability measurement
```

```python
rho2 = DensityMatrix(qc)
rho2.draw('latex',prefix='\\rho_2=')
```

```python
plot_state_city(qc)
```

```python
plot_state_hinton(qc)
```

```python
plot_state_paulivec(qc, title = 'Density matrix of the final state as linear combination of Pauli Gates')
```

```python
plot_bloch_multivector(qc)
```

```python
purityf = rho2.purity()
print(purityf)
```

```python
from qiskit.quantum_info import state_fidelity
```

```python
fidelity = state_fidelity(psi1,psi2)
print(fidelity)
```

```python
fidelity = state_fidelity(rho1,rho2)
print(fidelity)
```

```python
qc.measure(q[0],c[0])
```

```python
qc.draw('mpl')
```

```python
from qiskit import transpile
from qiskit.providers.aer import QasmSimulator
backend = QasmSimulator()
qc_compiled = transpile(qc,backend)
job = backend.run(qc_compiled,shots = 1024)
result = job.result()
counts = result.get_counts()
print(counts)
```

```python
from qiskit.visualization import plot_histogram
plot_histogram(counts)
```

Quick review: 
- R(x) give state 0 & 1 an equal finding probability
- R(y) relative phase is 0 and higher probability to find state 0
- R(z) give a global phase
