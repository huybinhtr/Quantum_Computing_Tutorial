```python

      
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Mon Jul 20 15:59:39 2020

@author: Vlad 
"""


# -*- coding: utf-8 -*-
"""
Spyder Editor

"""


import numpy as np


QubitNumber = 14
ElectronNumber=6



def EvenFlips(a):
    s=set()
    s.add((0,0))
    tmp=set()
    for i in range(QubitNumber):
        if a[i]=='Y' or a[i]=='X':
            if i%2==0:
                for element in s:
                    tmp.add((element[0]+1,element[1]+1))
                    tmp.add((element[0]-1,element[1]-1))
            else:
                for element in s:
                    tmp.add((element[0]+1,element[1]-1))
                    tmp.add((element[0]-1,element[1]+1))
            s=tmp
            tmp=set()
                
    if (0,0) in s:
        return True
    else:
        return False

    
    

def MoreThanDouble(a):
    count=0
    for i in range(QubitNumber):
        if a[i]=='Y' or a[i]=='X':
            count+=1
    if count>=5:
        return True
    else:
        return False


def Odd(a):
    result=-1
    for i in range(QubitNumber):
        if (a[i]=='Y'):
            result*=-1
    if result==1:
        return True
    else:
        return False

def SingleQubit(a):
    left, right=0,0
    for i in range(ElectronNumber):
        if a[i]=='X' or a[i]=='Y':
            left+=1
    for i in range(ElectronNumber,QubitNumber):
        if a[i]=='X' or a[i]=='Y':
            right+=1
    if right==1 and left==1:
        return True
    else:
        return False

def breaksSpin(a):
    spin=0
    for i in range(QubitNumber):
        if i<ElectronNumber:
            if i%2==0 and (a[i]=='Y' or a[i]=='X'):
                spin-=1
            elif a[i]=='Y' or a[i]=='X':
                spin+=1
        else:
            if i%2==0 and (a[i]=='Y' or a[i]=='X'):
                spin+=1
            elif a[i]=='Y' or a[i]=='X':
                spin-=1
            
    if spin!=0:
        return True
    else:
        return False  
    

    
def breaksNumber(a):
    right,left = 0,0
    for i in range(ElectronNumber):
        if (a[i]=='Y' or a[i]=='X'):
            left+=1
    for i in range(ElectronNumber,QubitNumber):
        if (a[i]=='Y' or a[i]=='X'):
            right+=1
    if left!=right:
        return True
    else:
        return False 

def Check(s):
    rep=0
    for i in range(8,10):
        if s[i]=='X' or s[i]=='Y':
            rep+=1
    if rep%2!=0:
        return False
    for i in range(0,2):
        if s[i]=='X' or s[i]=='Y':
            rep+=1
    for i in range(2,4):
        if s[i]=='X' or s[i]=='Y':
            rep+=1        
    for i in range(12,14):
        if s[i]=='X' or s[i]=='Y':
            rep+=1
    if rep%2!=0:
        return False
    for i in range(6,8):
        if s[i]=='X' or s[i]=='Y':
            rep+=1
    if rep%2!=0:
        return False
    for i in range(4,6):
        if s[i]=='X' or s[i]=='Y':
            rep+=1
    for i in range(10,12):
        if s[i]=='X' or s[i]=='Y':
            rep+=1        
    if rep%2!=0:
        return False
    return True
    






class SymmetrySets:
    
    def __init__(self):
        self.GoodPool=self.GenerateGoodPool()
        self.BadPool=self.GenerateBadPool()
        
        
    def GenerateBadPool(self):
        Set=['I','Y','X','Z']
        MyList=['']
        TmpList=[]
        i=0
        while i<QubitNumber:
            for element in MyList:
                for k in range(4):
                    tmp=element+Set[k]
                    TmpList.append(tmp)
            i+=1
            MyList=TmpList.copy()
            TmpList=[]
        for element in MyList:
            if Odd(element) and (MoreThanDouble(element) or  breaksNumber(element)\
                                        or breaksSpin(element) or \
                                            SingleQubit(element) or not Check(element)):
                TmpList.append(element)
        MyList=TmpList
        return MyList
    
                

                
    
    def GenerateGoodPool(self):
        Set=['I','Y','X','Z']
        MyList=['']
        TmpList=[]
        i=0
        while i<QubitNumber:
            for element in MyList:
                for k in range(4):
                    tmp=element+Set[k]
                    TmpList.append(tmp)
            i+=1
            MyList=TmpList.copy()
            TmpList=[]
        for element in MyList:
            if Odd(element) and not MoreThanDouble(element) and not \
                 SingleQubit(element) and not breaksNumber(element) and not breaksSpin(element) \
                    and Check(element):
                TmpList.append(element)
        MyList=TmpList
        return MyList   






def TestPool(Pool):
    ControlAlgebra=PauliAlgebra(Pool, Group=True, Algebra=True)
    if ControlAlgebra.CheckPool():
        print('The pool is complete')
        return True
    else:
        print('The pool is incomplete')
        return False




def CheckInseparability(Pool):
    Set=[Pool[0]]
    i=0
    while i<len(Set):
        for PoolElement in Pool:
            if PoolElement not in Set:
                if not CheckCommutivity(PoolElement, Set[i]):
                    Set.append(PoolElement)
        i+=1
    if len(Set)<len(Pool):
        return False
    else:
        return True
    



#Constructs a complete pool
def CompletePool():
    CompleteGenerators=[]
    for i in range(QubitNumber):
        if i==0:
            CompleteGenerators.append('Y')
        elif i==1:
            CompleteGenerators[0]='Z'+CompleteGenerators[0]
            CompleteGenerators.append('YI')
        else:
            for j in range(len(CompleteGenerators)):
                CompleteGenerators[j]='Z'+CompleteGenerators[j]
            CompleteGenerators.append('Y'+'I'*i)
            CompleteGenerators.append('I'+'Y'+'I'*(i-1))
    return CompleteGenerators





#The function checks if strings a and b anticommute and returns -1 in that case
def CheckCommutivity(a,b):
    result=1
    for i in range(QubitNumber):
        if  (a[i]=='X' and b[i]=='Y') or (a[i]=='X' and b[i]=='Z') or\
            (a[i]=='Y' and b[i]=='X') or (a[i]=='Y' and b[i]=='Z') or\
            (a[i]=='Z' and b[i]=='X') or (a[i]=='Z' and b[i]=='Y'):
            result*=-1
    if result==1:
        return True
    else:
        return False






#The function multiplies strings a and b and returns their product
def MultiplyStrings(a,b):
    result=''
    for i in range(QubitNumber):
        if a[i]=='X' and b[i]=='Y':
            result+='Z'
            
        if a[i]=='X' and b[i]=='Z':
            result+='Y'
		
        if a[i]=='Y' and b[i]=='X':
            result+='Z'
			
        if a[i]=='Y' and b[i]=='Z':
            result+='X'
			
        if a[i]=='Z' and b[i]=='X':
            result+='Y'
		
        if a[i]=='Z' and b[i]=='Y':
            result+='X'
			
        if a[i]=='Z' and b[i]=='Z':
            result+='I'
			
        if a[i]=='X' and b[i]=='X':
            result+='I'
			
        if a[i]=='Y' and b[i]=='Y':
            result+='I'
			
        if a[i]=='I':
            result+=b[i]
			
        if b[i]=='I' and a[i]!='I':
            result+=a[i]

    return result


    


	

        
       
        
class PauliAlgebra:

#The function to input the generator pool  or add elements to the existing pool
    def AddStrings(self):
        tmp = input('Insert Pauli Strings of length '+str(QubitNumber)\
                    +'or type exit to finish\n')
        while tmp!='exit':
            if tmp not in self.PauliStrings:
                self.Counter+=1
                self.PauliStrings.add(tmp)
            if tmp not in self.ProductStrings:
                self.ProductCounter+=1
                self.ProductStrings.add(tmp)
                if 'X' not in tmp and 'Y' not in tmp:
                    self.ZsubgroupCounter+=1
                    self.Zsubgroup.add(tmp)
            tmp = input()
        self.PauliStrings.sort()
        self.ProductStrings.sort()
        
#Constructs a class object from an inout pool provided as a parameter or from keyboard        
    def __init__(self,InputGenerators=None,Group=None,Algebra=None):
        if Group==None:
            Group=True
        if Algebra==None:
            Algebra=True
        if InputGenerators==None:
            self.PauliStrings=set()
            self.ProductStrings=set()
            self.Zsubgroup=set()
            self.ProductCounter=0
            self.ZsubgroupCounter=0
            self.Counter=0
            self.AddStrings()
        else:
            self.PauliStrings=set(InputGenerators)
            self.ProductStrings=set(InputGenerators)
            self.Counter=len(InputGenerators)
            self.ProductCounter=len(InputGenerators)
            self.Zsubgroup=set()
            self.ZsubgroupCounter=0
            for element in InputGenerators:
                if 'X' not in element and 'Y' not in element:
                    self.ZsubgroupCounter+=1
                    self.Zsubgroup.add(element)
        if Algebra==True:
            self.GeneratePauliStrings()
        if Group==True:
            self.GenerateGroup()



#Generates an algebra from a pool in self.PauliStrings and saves it there          
    def GeneratePauliStrings (self):
        MySet=self.PauliStrings.copy()
        tmpSet=self.PauliStrings.copy()
        helpSet=set()
        while len(tmpSet)!=0:
            for e1 in tmpSet:
                for e2 in MySet:
                    if not CheckCommutivity(e1, e2):
                        tmp=MultiplyStrings(e1,e2)
                        if tmp not in MySet:
                            helpSet.add(tmp)
            MySet=set.union(MySet,helpSet)
            tmpSet=helpSet
            helpSet=set()
        self.Counter=len(MySet)
        self.PauliStrings=MySet


    
    
    
#Prints all strings in self.PauliStrings 
    def printAlgebra(self):
        Dict={}
        for s in self.PauliStrings:
            Str=''
            for l in s:
                if l=='Y' or l=='X':
                    Str+='1'
                else:
                    Str+='0'
            if Str in Dict:
                Dict[Str].append(s)
            else:
                Dict[Str]=[s]
        for key in Dict.keys():
            Dict[key].sort()
        for key in sorted(Dict.keys(),key = lambda x: (str.count(x,'1'),x)):
            print(key, end=': ')
            for i in range(len(Dict[key])):
                if i!=len(Dict[key])-1:
                    print(Dict[key][i],end=', ')
                else:
                    print(Dict[key][i])
                
            

#Prints a product group from the set ProductStrings	    
    def printGroup(self):
        Dict={}
        for s in self.ProductStrings:
            Str=''
            for l in s:
                if l=='Y' or l=='X':
                    Str+='1'
                else:
                    Str+='0'
            if Str in Dict:
                Dict[Str].append(s)
            else:
                Dict[Str]=[s]
        for key in Dict.keys():
            Dict[key].sort()
        for key in sorted(Dict.keys(),key = lambda x: (str.count(x,'1'),x)):
            print(key, end=': ')
            for i in range(len(Dict[key])):
                if i!=len(Dict[key])-1:
                    print(Dict[key][i],end=', ')
                else:
                    print(Dict[key][i])

#Returns a subset of PauliStrings, 
#containig the biggest possible number of mutually anticommuting strings
    def BiggestAntiComSet(self,AntiComSet=None,MaxAntiComSet=None):
        Paulis=list(self.PauliStrings)
        if AntiComSet is None:
            AntiComSet=[]
        if MaxAntiComSet is None:
            MaxAntiComSet=[]    
        for i in range(self.Counter):
            if (len(AntiComSet)==0):
                AntiComSet.append(Paulis[i])
                MaxAntiComSet=self.BiggestAntiComSet(AntiComSet,MaxAntiComSet)
                if len(AntiComSet)>len(MaxAntiComSet):
                    MaxAntiComSet=AntiComSet.copy()
                AntiComSet.pop(-1)
            else:
                flag=0
                for j in range(len(AntiComSet)):
                    if CheckCommutivity(Paulis[i],AntiComSet[j]):
                        flag=1
                if (flag==0):
                    AntiComSet.append(Paulis[i])
                    MaxAntiComSet=self.BiggestAntiComSet(AntiComSet,MaxAntiComSet)
                    if len(AntiComSet)>len(MaxAntiComSet):
                        for i in range(len(AntiComSet)):
                            MaxAntiComSet=AntiComSet.copy()
                    AntiComSet.pop(-1)
        return MaxAntiComSet 
    




#Computes the nunmber of strings in PauliStrings, anticommuting with the string a
    def HowManyAntiCommute(self, a):
        num=0
        for Pauli in self.PauliStrings:
            if not CheckCommutivity(Pauli,a):
                num+=1
        return num
       


#Generates a product group from a pool in self.ProductStrings and saves it there  

    
    
    def GenerateGroup (self):
        self.ZsubgroupCounter=1
        self.Zsubgroup.add('I'* QubitNumber)
        tmpSet=set()
        tmpSet.add('I'* QubitNumber)
        helpSet=set()
        
        for Pauli in self.ProductStrings:
            if Pauli not in tmpSet:
                for el in tmpSet:
                    tmp=MultiplyStrings(Pauli,el)
                    helpSet.add(tmp)
                    if 'X' not in tmp and 'Y' not in tmp:
                        self.ZsubgroupCounter+=1
                        self.Zsubgroup.add(tmp)
                tmpSet=set.union(tmpSet,helpSet)
                helpSet=set()
        self.ProductCounter=len(tmpSet)
        self.ProductStrings=tmpSet
        
                    
                    
    def SimilarityTransformation(self, Operator):
        if not CheckIfOdd(Operator):
            return None
        
        tmpSet=set()
        for Pauli in self.ProductStrings:
            if not CheckCommutivity(Operator,Pauli):
                tmpSet.add(MultiplyStrings(Operator, Pauli))
            else:
                tmpSet.add(Pauli) 
        self.ProductStrings=tmpSet
        
        
        self.Zsubgroup=set()
        for element in self.ProductStrings:
            if 'X' not in element and 'Y' not in element:
                self.Zsubgroup.add(element)
         
        tmpSet=set()
        for Pauli in self.PauliStrings:
            if not CheckCommutivity(Operator, Pauli):
                tmpSet.add(MultiplyStrings(Operator, Pauli))
            else:
                tmpSet.add(Pauli)
        self.PauliStrings=tmpSet
        

        

#returns a subset of ProductStrings, 
#containing the strings connecting not more than n neighbouring qubits
    def ConnectivitySubSet(self,n):
        connectivitysubset=set()
        for element in self.ProductStrings:
            if CheckConnectivity(element)<=n and CheckIfOdd(element) :
                connectivitysubset.add(element)
        return connectivitysubset
        


#Checks if the generator pool was complete
    def CheckPool(self):
        if self.CheckGroup() and self.CheckAlgebraSize():
            return True
        else:
            return False
        
        
#Checks if the group has the right form (size of zerosubgroup + size of the group + flippings)     
    def CheckGroup(self):
        if self.ZsubgroupCounter==2**(QubitNumber-2) and\
            self.ProductCounter==2**(2*QubitNumber-2):
                Str=''
                my_array=np.zeros((2**QubitNumber-1,), dtype=int)
                for element in self.ProductStrings:
                    if CheckIfOdd(element):
                        for letter in element:
                            if letter=='I' or letter=='Z':
                                Str=Str+'0'
                            if letter=='Y' or letter=='X':
                                Str=Str+'1'
                        my_array[int(Str,2)-1]=1
                        Str=''
                if 0 not in my_array:
                    return True
                else:
                    return False
        else:
            return False
  
#Checks if the algebra size is correct                
    def CheckAlgebraSize(self):
        if len(self.PauliStrings)==2**(QubitNumber-2)*(2**(QubitNumber-1)+1):
            return True
        else:
            return False
#Checks if the algebra produces all fleepings of the qubits 
#(necessary condition of completeness)         
    def CheckFlippings(self,start=0):
        Str=''
        my_array=np.zeros((2**QubitNumber-1,), dtype=int)
        for element in self.PauliStrings:
            for letter in element:
                if letter=='I' or letter=='Z':
                    Str=Str+'0'
                if letter=='Y' or letter=='X':
                    Str=Str+'1'
            my_array[int(Str,2)-1]=1
            Str=''
        if 0 not in my_array:
            return True
        else:
            return False





#Returns the nuber of neighbouring qubits, connected in the string a
def CheckConnectivity(a):
    Connectivity=0
    i=0
    while i<len(a) and a[i]=='I' :
        i+=1
    if i==len(a):
        return Connectivity
    else:
        Connectivity=1
    for j in range(i+1,len(a)):
        if a[j]!='I':
            Connectivity=j-i+1
    return Connectivity
        

#Returns a Python set of odd Paulis, with connectivity <= n
def OddConnectivityNotBigger(n):
    Alphabet=['I','Y','X','Z']
    
    MyList=['']
    TmpList=[]

    for i in range(QubitNumber):
        for element in MyList:
            for k in range(4):
                TmpList.append(element+Alphabet[k])
        MyList=TmpList
        TmpList=[]
    
    MySet=set()
    for element in MyList:
        if CheckIfOdd(element) and CheckConnectivity(element)<=n:
            MySet.add(element)

    return MySet


#Returns a Python set of odd Paulis, with connectivity == n
def OddConnectivityEqual(n):
    Alphabet=['I','Y','X','Z']
    
    MyList=['']
    TmpList=[]

    for i in range(QubitNumber):
        for element in MyList:
            for k in range(4):
                TmpList.append(element+Alphabet[k])
        MyList=TmpList
        TmpList=[]
    
    MySet=set()
    for element in MyList:
        if CheckIfOdd(element) and CheckConnectivity(element)==n:
            MySet.add(element)

    return MySet

#Swaps qubits i and j in a Pauli a
def Swap(i,j,a):
    if i>j:
        tmp=i
        i=j
        j=tmp
    return a[:i]+a[j]+a[i+1:j]+a[i]+a[j+1:]



#Searches for random minimal complete pools (most useful for big qubits numbers) 
def FindRandomCompletePools(Operator = None):
    if Operator==None:
        Operator='I'*QubitNumber
        

    
    Group=PauliAlgebra(CompletePool(),Algebra=False)
    SubstractSet=Group.ProductStrings.copy()
    AllOddStrings=OddConnectivityNotBigger(QubitNumber)
    
    AllRotators=set.difference(AllOddStrings,SubstractSet)
    tmpSet=AllRotators.copy()
    
    Sets=[Group.ProductStrings.copy()] 
    while len(AllRotators)>0:
        for Operator in AllRotators:
            Group.SimilarityTransformation(Operator)
            tmp=Group.ProductStrings.copy()
            if set.intersection(tmpSet,tmp)==set():
                tmpSet.remove(Operator)
                Group.SimilarityTransformation(Operator)
                flag=1
                for aset in Sets:
                    if aset==tmp:
                        flag=0
                if flag==1:
                    Sets.append(tmp)
            else:
                tmpSet.remove(Operator)
                tmpSet=set.difference(tmpSet,tmp)
                Group.SimilarityTransformation(Operator)
                flag=1
                for aset in Sets:
                    if aset==tmp:
                        flag=0
                if flag==1:
                    Sets.append(tmp)
                break
        AllRotators=tmpSet
        tmpSet=AllRotators.copy()
            
    

    SampleSets=[]
    for aset in Sets:
        SampleSets.append(list(aset))
        
    CompletePools=[]
    CompleteCounter=0 
    while CompleteCounter<10000:
        MyList=SampleSets[np.random.randint(0,len(SampleSets))]
        element=[]
        for i in range(2*QubitNumber-2):
            element.append(MyList[np.random.randint(0,len(MyList))])
        if CheckInseparability(element):
            tmp=PauliAlgebra(element,Group=True,Algebra=False)
            if tmp.CheckGroup():
                CompleteCounter+=1
                CompletePools.append(element)
                if CompleteCounter % 1 ==0:
                    for element in CompletePools[CompleteCounter-1:CompleteCounter]:
                       print(element)

    return CompletePools



import random

S=SymmetrySets()
   


CP=PauliAlgebra(CompletePool(),Group=True,Algebra=False)
FullSet=CP.ProductStrings.copy()
FullGroup=CP.ProductStrings.copy()

def printStarters():
    starters = set.intersection(set(S.GoodPool), set(FullSet))
    random.shuffle(list(starters))
    print(starters)

#l=set.intersection(set(S.GoodPool),set(FullSet))
#random.shuffle(list(l))
#random.shuffle(list(l))
#print(l)






print(len(FullSet))
print(len(FullGroup)) 

      



printer = printStarters()

for el in CP.ProductStrings:
    if not Odd(el) or not Check(el) or not EvenFlips(el):
        FullSet.remove(el)
        
for el in CP.ProductStrings:
    if not Check(el) or not EvenFlips(el):
        FullGroup.remove(el)
        

        
print(len(FullSet))
print(len(FullGroup))  
                           
MyList=list(FullSet)        
#print(MyList)
CompleteCounter=0 
while True:
    element= ['IZZIYYZIIIIIXY', 'IIYYZZIZZIIZXY', 'IZXIIXIIIIXIZY', 'ZZZXYZIZZZYIZY', 'IIYIZXZZIZZYYI', 'IIIYYIZZIZZXYI', 'IZZZYYZIIIXYII', 'ZZXXZZZZIZXYII', 'ZIYXIIIIXXZZII', 'YYZZZZIIIZZZXY', 'XYIIIZZZZIYYII', 'ZIIYIYIIIZIXZY', 'ZZXZYZZIIIYIYI', 'IIZZYXIIXXIZII', 'YZIXZZIZYYZIII', 'ZYYZIZIIXYIZII', 'YYIZIZIIXYIIII', 'XZIXZZZIZIIZXY', 'IYYIIZZZIIIIXY', 'YIZIZYZZZZXIZY', 'ZYZZYZZIIZXZZY', 'YZIZZYIIIZIXYI', 'ZYIIXIZIZZZYYI', 'YZIXZIIZIIXXII', 'ZXXZIZIZIZXYII', 'IXZIIXZIZIIXZY', 'YZIZYIIIZIXIYI']

    for i in range(55):
        element.append(MyList[np.random.randint(0,len(MyList))])
    if CheckInseparability(element):
        print('A')
        tmp=PauliAlgebra(element,Group=True,Algebra=False)
        if len(tmp.ProductStrings)==len(FullGroup):
            print('B')
            print(element)
            tmp.GeneratePauliStrings()
            print(FullSet.difference(tmp.PauliStrings))
            if len(tmp.PauliStrings)==len(FullSet):            
                MyCompletePool=element
                print(MyCompletePool)
                break
      



                   



  
       
```
