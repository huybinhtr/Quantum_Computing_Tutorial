## Bell State - Maximally entanglement

```python
from qiskit import QuantumRegister, QuantumCircuit
q = QuantumRegister(2)
qc0 = QuantumCircuit(q)
qc1 = QuantumCircuit(q)
qc2 = QuantumCircuit(q)
qc3 = QuantumCircuit(q) # each has two qubits
```

```python
from qiskit.quantum_info import Statevector
```

```python
qc0.h(q[0])
qc0.cx(q[0],q[1]) # control on qubit 0 and target on qubit 1
qc0.draw('mpl')
```

```python
phi_plus = Statevector(qc0)
phi_plus.draw('latex')
```

```python
qc1.h(q[0])
qc1.cx(q[0],q[1])
qc1.z(q[0])
qc1.draw('mpl')
```

```python
phi_minus = Statevector(qc1)
phi_minus.draw('latex')
```

```python
qc2.h(q[0])
qc2.cx(q[0],q[1])
qc2.x(q[1])
qc2.draw('mpl')
```

```python
psi_plus = Statevector(qc2)
psi_plus.draw('latex')
```

```python
qc3.h(q[0])
qc3.cx(q[0],q[1])
qc3.z(q[0])
qc3.x(q[1])
qc3.draw('mpl')
```

```python
psi_minus = Statevector(qc3)
psi_minus.draw('latex')
```

```python
display(qc0.draw('mpl'),qc1.draw('mpl'),qc2.draw('mpl'),qc3.draw('mpl'))
```

```python
plot_bloch_multivector(qc2)
```

```python
plot_bloch_multivector(qc3)
```

As you see we have no results because we can not define the qs for qubit 0 and 1 separately (decompse) so that proove for the engtanglement

```python
print(phi_plus.purity()) # purity appoach to one is well define 
```

They are all pure state, and maximal diagonal state that means their reduce density states are completly mixed

```python
from qiskit.quantum_info import DensityMatrix
```

```python
rho_plus = DensityMatrix(qc0)
rho_plus.draw('latex',prefix = '\\rho_{AB}=')
```

```python
from qiskit.quantum_info import partial_trace
```

```python
rho_A = partial_trace(rho_plus,[0])
rho_B = partial_trace(rho_plus,[1])
```

```python
rho_A.draw('latex',prefix='\\rho_A=')
```

```python
rho_B.draw('latex',prefix='\\rho_B=')
```

```python
rho_A.purity()
```

```python
rho_B.purity()
```

```python
# these two states are completely mixed 
```
